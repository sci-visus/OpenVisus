# Show single sice

```python
url='http://atlantis.sci.utah.edu:80/mod_visus?dataset=chess-zip&cached=1'

# need to know how to display the data
palette="Viridis256"
palette_range=-0.017141795,0.012004322 

def MyApp(doc): 
    view=Slice(doc,show_options=["palette","timestep","field","direction","offset"])
    view.setDataset(url)   
    view.setDirection(2)
    view.setPalette(palette)
    view.setPaletteRange(palette_range)
    view.setTimestep(view.getTimestep())
    view.setField(view.getField())         
    doc.add_root(view.layout)
ShowApp(MyApp)
```

# Show multiple slices

```python
def MyApp(doc):
    view=Slices(doc,num_views=2)
    view.setDataset(url)
    view.setPalette(palette)
    view.setPaletteRange(palette_range)
    view.setTimestep(view.getTimestep())
    view.setField(view.getField())       
    doc.add_root(view.layout)
ShowApp(MyApp)
```
